```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image_color1 = mpimg.imread("/content/image_lane_c.jpg")
image_color2 = cv2.imread("/content/new_test_image_3.jpg")
```

```python
plt.imshow(image_color1)
height1, width1 = image_color1.shape[:2]
print("Height image 1 : "+str(height1))
print("Width image 1 : "+str(width1))
```

```python
plt.imshow(image_color2)
height2, width2 = image_color2.shape[:2]
print("Height image 2 : "+str(height2))
print("Width image 2 : "+str(width2))
```

# Convert to Greyscale

```python
image_gray_1 = cv2.cvtColor(image_color1, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray_1, cmap='gray')
print(image_gray_1.shape)
```

```python
image_gray_2 = cv2.cvtColor(image_color2, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray_2, cmap='gray')
print(image_gray_2.shape)
```

# Perform Region of Interest Masking

```python
# Selection points of the region of interest (ROI)
ROI_1 = np.array([[(0, height1), (320, 300), (500, 300), (width1, height1)]], dtype=np.int32)
ROI_2 = np.array([[(0, height2), (320, 300), (500, 300), (width2, height2)]], dtype=np.int32)

# define a blank image with all zeros (ie: black) 
blank_1 = np.zeros_like(image_gray_1)
blank_2 = np.zeros_like(image_gray_2)
```

```python
mask_1 = cv2.fillPoly(blank_1, ROI_1, 255)
mask_2 = cv2.fillPoly(blank_2, ROI_2, 255)

masked_image_1 = cv2.bitwise_and(image_gray_1, mask_1)
masked_image_2 = cv2.bitwise_and(image_gray_2, mask_2)
```

```python
plt.imshow(masked_image_1, cmap='gray')
```

```python
plt.imshow(masked_image_2, cmap='gray')
```
