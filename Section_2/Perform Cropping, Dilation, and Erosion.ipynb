```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline

image = mpimg.imread("/content/Self_Driving_Car.jpg")
plt.imshow(image)
```

```python
height, width = image.shape[:2]
print("width : "+str(width))
print("height : "+str(height))
```

# Perform Cropping

```python
# Top left coordinates of the desired cropped area (w0, h0)
w0 = int(width * 0.5)
h0 = int(height * 0.5)

# Bottom right coordinates of the desired cropped area (w1, h1)
w1 = int(width * 1)
h1 = int(height * 1)

Image_cropped = image[h0:h1, w0:w1]
plt.imshow(Image_cropped)
```

# Perform Dilation and Erosion Operations

```python
new_image = mpimg.imread("/content/Speed_Sign_View_2.jpg")
kernel = np.ones((3,3), np.uint8)
image_dilation = cv2.dilate(new_image, kernel, iterations = 3)
plt.imshow(image_dilation)
```

```python
image_erosion = cv2.erode(new_image, kernel, iterations = 3)
plt.imshow(image_erosion)
```
