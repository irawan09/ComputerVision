```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image = plt.imread('/content/Self_driving_car.jpg')
plt.imshow(image)
```

```python
height, width = image.shape[:2]
print("height :"+str(height))
print("width :"+str(width))
```

## Perform Rotations

```python
M_rotation = cv2.getRotationMatrix2D((width/2, height/2), 90, 0.5)
plt.imshow(M_rotation)
```

```python
rotated_image = cv2.warpAffine(image, M_rotation, (width, height))
plt.imshow(rotated_image)
```

# Perform Translations

```python

Translational_Matrix = np.float32([[1, 0, 300],
                                  [0, 1, 50]])

translated_image = cv2.warpAffine(image, Translational_Matrix, (width, height))
plt.imshow(translated_image)
```

# Perform Resizing

```python
resized_image = cv2.resize(image, None, fx=4, fy=4, interpolation = cv2.INTER_CUBIC)
plt.imshow(resized_image)
```
