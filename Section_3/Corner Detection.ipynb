```python
import cv2 
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image_color = plt.imread('/content/chessboard.jpg')
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray)
```

```python
corners = cv2.cornerHarris(image_gray, 5, 3, 0.1)
plt.imshow(corners, cmap='gray')
```

```python
corners_dilate = cv2.dilate(corners, np.ones((8,8),np.uint8), iterations = 1)
plt.imshow(corners_dilate, cmap='gray')
```

```python
detector = corners_dilate > 0.01 * corners_dilate.max()
```
