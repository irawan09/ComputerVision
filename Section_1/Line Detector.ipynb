## Color Selection (Line Detector)

## Import all Necessary Libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

## Read the image and Display it

```python
image_color = mpimg.imread('/content/bg.jpg')
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray)
```

```python
image_gray.shape
```

## Find the white color from Grayscale Image

We will change the other color into black (0) and leave the white color only in the picture (255)

```python
image_copy = np.copy(image_gray)
image_copy.shape
```

```python
image_copy[(image_copy[:, :] < 200)] = 0
```

```python
plt.imshow(image_copy, cmap='gray')
plt.show
```

## Line detection in Color Image

```python
image_copy_color = np.copy(image_color)
image_copy_color.shape
```

removing the other color by its own RGB channel

```python
image_copy_color[(image_copy_color[:, :, 0] < 200) | (image_copy_color[:, :, 1] < 200) | (image_copy_color[:, :, 2] < 200)] = 0
```

```python
plt.imshow(image_copy_color, cmap='gray')
plt.show
```
