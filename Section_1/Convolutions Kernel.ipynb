# Convolutional Kernel

In image processing, a kernel, convolution matrix, or mask is a small matrix used for blurring, sharpening, embossing, edge detection, and more. This is accomplished by doing a convolution between the kernel and an image. [ Source](https://en.wikipedia.org/wiki/Kernel_(image_processing))

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
import cv2
import numpy as np
```

```python
image = cv2.imread('/content/drive/MyDrive/ComputerVision/car.jpg')

plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image)
```

## Sharpening Kernel

```python
sharp_kernel_1 = np.array([[0, -1, 0],
                         [-1, 5, -1],
                         [0, -1, 0]])


sharp_kernel_2 = np.array([[-1, -1, -1],
                         [-1, 9, -1],
                         [-1, -1, -1]])

sharpened_image = cv2.filter2D(gray_image, -1, sharp_kernel_1)
plt.imshow(sharpened_image)
```

```python
sharpened_image = cv2.filter2D(gray_image, -1, sharp_kernel_2)
plt.imshow(sharpened_image)
```

## Blurring Kernel

```python
box_blurr_kernel = np.ones((3,3))
```

```python
blurred_image = cv2.filter2D(gray_image, -1, box_blurr_kernel)

plt.imshow(blurred_image)
plt.title("Kernel before Normalize")
```

## Normalize Kernel

```python
normalize_blurr_kernel = box_blurr_kernel * (1/9)

normalized_blurred_image = cv2.filter2D(gray_image, -1 , normalize_blurr_kernel)

plt.imshow(normalized_blurred_image)
plt.title("Kernel After Normalize")
```

Extending the convolutional kernel size :

```python
big_blurr_kernel = np.ones((8,8)) * (1/64)

extending_blurr_kernel = cv2.filter2D(gray_image, -1, big_blurr_kernel)
plt.imshow(extending_blurr_kernel)
plt.title('Extending Blurr Kernel')
```
