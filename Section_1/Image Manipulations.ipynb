# Image Manipulations (BGR/RGB/HSV/GRAY/SPLIT/MERGE)

## HSV (Hue Saturation Value)

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_image)
```

```python
plt.imshow(hsv_image[:, :, 0])
plt.title('Hue Channel')
```

```python
plt.imshow(hsv_image[:, :, 1])
plt.title('Saturation Channel')
```

```python
plt.imshow(hsv_image[:, :, 2])
plt.title('Value Channel')
```

## Split and Merge RGB channels

```python
B, G, R = cv2.split(image)

plt.imshow(B)
plt.title('Blue Channel')
```

```python
plt.imshow(R)
plt.title('Red Channel')
```

```python
plt.imshow(G)
plt.title('Green Channel')
```

Manipulates the others channel into zero

```python
zeros = np.zeros(image.shape[:2], dtype = "uint8")
merge_blue = cv2.merge([B, zeros, zeros])

plt.imshow(merge_blue)
plt.title("Red channel (Manipulated)")
```

```python
merge_image = cv2.merge([B, G, R])

plt.imshow(merge_image)
plt.title('Merged Image')
```

```python
green_color = cv2.merge([G,R,B])

plt.imshow(green_color)
plt.title('Shuffle Color Image')
```

```python
red_color = cv2.merge([R,G,B])

plt.imshow(red_color)
plt.title('Shuffle Color Image')
```
